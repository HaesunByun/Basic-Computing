- 기획자 정보 (학과(부)/학번/성명): 조청유과/1234-56789/김허니
- e-mail: xxx@snu.ac.kr
- 난이도: 중


## 나: 요구사항명세서


게임 이론에 등장하는 최후통첩 게임과 독재자 게임을 할 수 있도록 하는 프로그램을 제작해주세요.<br>

이 게임에 참가하는 플레이어는 A, B, 그리고 컴퓨터입니다.
게임이 시작하면 안내문이 나온 후, 무작위로 최후통첩 게임과 독재자 게임 중 하나가 선택되어 진행됩니다.

-게임을 반복해서 진행할 수 있도록 한다.

### <최후통첩 게임>

최후통첩 게임에는 두 명의 제안자와 한 명의 결정자가 있습니다. 게임이 시작되면, 먼저 결정자 한 명을 무작위로 골라주세요. 첫 번째 제안자가 100달러를 세 명이 어떻게 나누어 가질지 A, B, 컴퓨터 순서로 입력해주세요.두 번째 제안자가 100달러를 세 명이 어떻게 나누어 가질지 A, B, 컴퓨터 순서로 입력해주세요. 마지막으로 결정자가 두 개의 제안을 보고, 어떤 제안을 받아들일지 선택해주세요. 상금은 결정자가 선택한 대로 나누어집니다.

컴퓨터가 제안자가 되면, 나누어 가지는 비율을 무작위로 고르도록 해주세요. 컴퓨터가 결정자가 되면, 자신에게 더 많은 돈을 주겠다고 한 제안을 고르도록 해주세요.

### <독재자 게임>
독재자 게임이 시작되면 먼저 무작위로 한 명의 독재자를 골라주세요. 독재자가 100달러를 세 명이 어떻게 나누어 가질지 A, B, 컴퓨터 순서로 입력해주세요. 상금은 독재자가 정한 대로 나누어 집니다.

컴퓨터가 독재자가 되면, 다음과 같은 기준으로 상금을 분배하도록 해주세요.
i) 이전 라운드까지 상금 합계에서 컴퓨터보다 상금이 적은 사람이 없는 경우, 컴퓨터가 100달러를 모두 갖도록 해주세요.
ii) 그 외의 경우, 자신보다 상금이 적은 모두에게 50달러를 주고, 나머지를 컴퓨터가 갖도록 해주세요.

## 다: 세부 요구사항

### 요구사항 1

게임에 참가하는 플레이어는 A, B, 그리고 컴퓨터입니다. 게임이 시작하면 먼저 다음과 같은 안내문이 1초의 시간 간격을 두고 출력시켜 주세요.

```
"변형된 최후통첩 게임에 참여하신 여러분을 환영합니다.

여러분은 1회와 20회 사이의 게임에 참여하게 될 것입니다. 게임의 횟수는 무작위로 선정됩니다.

각 게임은 최후통첩 게임 또는 독재자 게임이며, 게임에 대한 설명은 게임이 시작된 후 자세히 들으실 수 있습니다.

여러분의 목표는, 각 회에 걸려있는 100달러의 상금 중 최대한 많은 상금을 얻어 여러분의 경쟁자와 컴퓨터를 이기는 것입니다.

그러면 첫 번째 게임을 선정하겠습니다."
```








### 요구사항 2
이후 무작위로 '최후통첩 게임'과 '독재자 게임' 중 하나를 선택하여 진행하며, 다음과 같은 문구를 출력시켜 주세요.

```
"제 n(게임 횟수에 따라 숫자를 달리해주세요)회 게임은 (게임 이름)입니다."
```

### 요구사항 3
게임에 대한 설명은 처음으로 그 게임을 진행할 때만 출력되도록 해주세요. 즉, 최후통첩 게임에 대한 설명은 최후통첩 게임을 처음 진행한 회에, 독재자 게임에 대한 설명은 독재자 게임을 처음 진행한 회에 출력되도록 해주세요. 게임에 대한 설명을 출력한 이후, 숙지할 시간 30초를 주고 다음으로 넘어가 주세요.

### 요구사항 4
이후 게임을 진행해 주시고, 게임이 마무리 되면 각 플레이어가 얻은 상금을 표시해 주세요.

### 요구사항 5
무작위로 1회부터 20회 사이의 게임을 반복해 주시고(1회 만에 게임이 끝날 수도, 20회까지 게임이 진행될 수도 있음), 게임이 끝나면 총 상금과 최종 승자를 표시해 주세요.
