## 4.(함께) 상관관계 확인
- 상관관계의 정도는 상관계수로 정량화해서 파악할 수 있다.
- 상관계수(Correlation Coefficient)는 corrcoef( )나 corr()를 통해 확인한다.
  - .corrcoef(x, y): 상관 계수행렬을 반환한다.
  - .corr(): 모든 행렬들간 상관계수의 행렬을 보여준다.

```
- 1.0 <= r < -0.7: 매우 강한 음의 상관관계
- 0.7 <= r < -0.3: 강한 음의 상관관계
- 0.3 <= r < -0.1: 약한 음의 상관관계
- 0.1 <= r <= 0.1: 상관 관계 없거나 무시할만 함
- 0.1 < r <= +0.3: 약한 양의 상관관계
- 0.3 < r <= 0.7: 강한 양의 상관관계
- 0.7 < r <= 1.0: 매우 강한 양의 상관관계
```
