### ì¹´í…Œê³ ë¦¬ë³„ íšŒê·€ì„ ì´ ìˆëŠ” ì‚°í¬ë„ ê·¸ë¦¬ê¸°: lmplot()

### íˆìŠ¤í† ê·¸ë¨ê³¼ ì»¤ë„ë°€ë„ ê·¸ë˜í”„: displot()

- ê¸°ë³¸ê°’ìœ¼ë¡œ íˆìŠ¤í† ê·¸ë¨ê³¼ ì»¤ë„ ë°€ë„ í•¨ìˆ˜ë¥¼ ê·¸ë˜í”„ë¡œ ë³´ì—¬ì¤€ë‹¤.
