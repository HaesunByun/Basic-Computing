#### 1.생존여부에 따른 객실등급 분포

- 생존여부(Survived, Dead)를 인덱스로 갖는 데이터프레임을 만들자.
- titanic 데이터프레임에서 Survived 열의 값이 1인 행이 생존자이다.
  

````python
s = titanic['Survived']==1  
titanic[s]
````

- titanic 데이터프레임에서 Survived 열의 값이 0인 행이 비생존자이다.
- 😄 비생존자에 대한 조건식의 결과를 변수 d로 저장하시오.


- value_counts()함수를 이용하여 Pclass(객실등급)별로 생존자 수를 count하여 survived라는 시리즈를 만든다.

````python
survived = titanic[s]['Pclass'].value_counts()
survived
````

- value_counts()의 경우 survived를 출력한 결과에서와 같이 내림차순으로 정렬되어 만들어지므로 이를 sort_index()를 이용하여 인덱스 기준으로 정렬한다. 
- 여기서의 인덱스는 Pclass인 객실등급이다.

````python
survived.sort_index(ascending=True, inplace=True)
survived
````

- 😄 Pclass(객실등급)별로 비생존자 수를 count하여 dead라는 시리즈를 만드시오.

- survived, dead 시리즈를 데이터프레임 df로 만든다.
- 이때 인덱스는 'Survived', 'Dead'로 준다.

```
df= pd.DataFrame([survived, dead], index = ['Survived', 'Dead'])
df
```


- 지금까지 한 작업을 함수로 만들어 보자.


````python
s = titanic['Survived']==1    
d = titanic['Survived']==0

def bar_chart():
    survived = titanic[s]['Pclass'].value_counts()
    dead = titanic[d]['Pclass'].value_counts()
    survived.sort_index(ascending=True, inplace=True)

    df= pd.DataFrame([survived, dead], index = ['Survived', 'Dead'])
    df.plot(kind='barh', stacked = True, figsize=(10,2))
    plt.title(f'Survived & Dead by Pclass')
    plt.xlabel('Count')        
    plt.ylabel('Survived & Dead')
    plt.show()

bar_chart()
````

- 함수로 만들어 놓은 이유는 'Pclass'이외의 다른 칼럼의 값을 넣어서 동일한 형태의 그래프를 쉽게 그릴 수 있다. 
- 함수에 전달인자를 넣어서 전달인자에 따라 그래프가 그려지도록 수정해보자.
  - 변수 feature에 전달인자를 받아, 수평누적막대 그래프를 그리도록 수정한다.
  - 즉, 모든 'Pclass'를 feature로 바꿔준다.

````python
def bar_chart(feature):
    
    survived = titanic[s][feature].value_counts()
    dead = titanic[d][feature].value_counts()
    survived.sort_index(ascending=True, inplace=True)

    df1= pd.DataFrame([survived, dead], index = ['Survived', 'Dead'])
    df1.plot(kind='barh', stacked = True, figsize=(10,2))
    plt.title(f'Survived & Dead by {feature}')
    plt.ylabel('Survived & Dead')
    plt.show()
````

- 'Pclass'(객실등급)을 전달인자로 줘서 그래프를 그려보자.
- 이 그래프를 보고 무엇을 설명할 수 있는지 말해보자.

````python
bar_chart('Pclass')
````
