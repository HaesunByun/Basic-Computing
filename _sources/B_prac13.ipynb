## -- 붓꽃 데이터 활용 --

- sns.load_dataset()으로 붓꽃(아이리스) 데이터를 불러와 그래프를 그려보자.
- 붓꽃 데이터는 시각화에서 유명한 데이터라 seaborn에 데이터셋으로 들어가 있다. 이 외에도 titanic, tips, penquins, flights, diamonds 등이 있다.
- .head()로 붓꽃 데이터의 다섯 개 행의 정보를 확인해보자.
- 붓꽃 데이터 열의 의미는 다음과 같다.
  - sepal_length : 꽃받침 길이
  - sepal_width : 꽃받침 너비
  - petal_length : 꽃잎 길이
  - petal_width : 꽃잎 너비
  - species : 아이리스 종류(setosa, versicolor, virginica)

```python
import seaborn as sns
iris = sns.load_dataset("iris") #iris(붓꽃)데이터셋
iris.head()
```

### (함께)pairplot 그래프: .pairplot()
- **pairplot그래프**는 데이터프레임의 열(변수)을 두 개씩 짝을 지을 수 있는 모든 조합에 대해 나타낸다.
- 만들어진 짝의 개수만큼 화면을 그리드로 나눈다.
- 같은 변수끼리 짝을 이루는 경우에는 **커널밀도 그래프**로 나타내고, 서로 다른 변수 간에는 **산점도 그래프**를 나타낸다.
- 옵션 `hue`는 카테고리별로 보여준다.

😄 붓꽃 데이터의 붓꽃 종류별 4가지(꽃받침, 꽃잎의 길이와 너비) 조합에 대해 pairplot()을 그려보자.이때 품종별로 색상을 보여주자.

### (함께)FacetGrid 그래프: .FacetGrid()
- **FacetGrid그래프**는 행, 열 방향으로 서로 다른 조건을 적용하여 여러 개의 서브 플롯을 만든다.
- map() 함수를 이용하여 각 서브 플롯에 적용할 그래프 종류를 전달한다.
- 😄 붓꽃 종류(품종)별로 열방향으로 그리드를 나눠 히스토그램을 그려보자. 이때 품종별로 색상을 보여주자.
