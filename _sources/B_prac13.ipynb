- ì§€ê¸ˆê¹Œì§€ í•œ ì‘ì—…ì„ í•¨ìˆ˜ë¡œ ë§Œë“¤ì–´ ë³´ì.


````python
s = titanic['Survived']==1    
d = titanic['Survived']==0

def bar_chart():
    survived = titanic[s]['Pclass'].value_counts()
    dead = titanic[d]['Pclass'].value_counts()
    survived.sort_index(ascending=True, inplace=True)

    df= pd.DataFrame([survived, dead], index = ['Survived', 'Dead'])
    df.plot(kind='barh', stacked = True, figsize=(10,2))
    plt.title(f'Survived & Dead by Pclass')
    plt.xlabel('Count')        
    plt.ylabel('Survived & Dead')
    plt.show()

bar_chart()
````

- í•¨ìˆ˜ë¡œ ë§Œë“¤ì–´ ë†“ì€ ì´ìœ ëŠ” 'Pclass'ì´ì™¸ì˜ ë‹¤ë¥¸ ì¹¼ëŸ¼ì˜ ê°’ì„ ë„£ì–´ì„œ ë™ì¼í•œ í˜•íƒœì˜ ê·¸ë˜í”„ë¥¼ ì‰½ê²Œ ê·¸ë¦´ ìˆ˜ ìˆë‹¤. 
- í•¨ìˆ˜ì— ì „ë‹¬ì¸ìë¥¼ ë„£ì–´ì„œ ì „ë‹¬ì¸ìì— ë”°ë¼ ê·¸ë˜í”„ê°€ ê·¸ë ¤ì§€ë„ë¡ ìˆ˜ì •í•´ë³´ì.
  - ë³€ìˆ˜ featureì— ì „ë‹¬ì¸ìë¥¼ ë°›ì•„, ìˆ˜í‰ëˆ„ì ë§‰ëŒ€ ê·¸ë˜í”„ë¥¼ ê·¸ë¦¬ë„ë¡ ìˆ˜ì •í•œë‹¤.
  - ì¦‰, ëª¨ë“  'Pclass'ë¥¼ featureë¡œ ë°”ê¿”ì¤€ë‹¤.

````python
def bar_chart(feature):
    
    survived = titanic[s][feature].value_counts()
    dead = titanic[d][feature].value_counts()
    survived.sort_index(ascending=True, inplace=True)

    df1= pd.DataFrame([survived, dead], index = ['Survived', 'Dead'])
    df1.plot(kind='barh', stacked = True, figsize=(10,2))
    plt.title(f'Survived & Dead by {feature}')
    plt.ylabel('Survived & Dead')
    plt.show()
````

- 'Pclass'(ê°ì‹¤ë“±ê¸‰)ì„ ì „ë‹¬ì¸ìë¡œ ì¤˜ì„œ ê·¸ë˜í”„ë¥¼ ê·¸ë ¤ë³´ì.
- ì´ ê·¸ë˜í”„ë¥¼ ë³´ê³  ë¬´ì—‡ì„ ì„¤ëª…í•  ìˆ˜ ìˆëŠ”ì§€ ë§í•´ë³´ì.

````python
bar_chart('Pclass')
````
