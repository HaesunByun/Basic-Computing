### (참고) if vs. list vs. dict 코드 비교

<font color='blue'>`if문`</font>
```python
if문을 활용하였을 경우 해당 운동종목마다 칼로리를 계산하는 코드를 각각 작성해줘야 한다.
```
```
weight = input('체중(kg): ')
time = input('운동시간(m): ')
exercise = input('운동종목(걷기, 조깅, 계단오르기, 수영): ')


weight = int(weight)
time = int(time)
if (exercise == '걷기') :
  calories = ((4 * 3.5 * weight) / 200) * time
elif (exercise == '조깅') :
  calories = ((7 * 3.5 * weight) / 200) * time
elif (exercise == '계단오르기') :
  calories = ((5 * 3.5 * weight) / 200) * time
elif (exercise == '수영') :
  calories = ((9 * 3.5 * weight) / 200) * time
else :
  print('포함되지 않은 운동종목입니다.')  

print(f'{exercise} {time}분은 {calories}kcal 소모')
```

<font color='blue'>`list`</font>
```python
list의 경우 두 개의 리스트에 운동종목과 Met가 정확한 순서로 들어가 있어야 한다.
```

```python
ex_list=['걷기', '조깅', '계단오르기', '수영']
ex_Mets=[4,7,5,9]

weight = int(input('체중(kg): '))
time = int(input('운동시간(m): '))
exercise = input(f'운동종목 {ex_list}: ')

if exercise in ex_list:
  num = ex_list.index(exercise)
  calories = ((ex_Mets[num] * 3.5 * weight) / 200) * time
  print(f'{ex_list[num]} {time}분은 {calories}kcal 소모')
else:
  print('포함되지 않은 운동종목입니다.')
```
