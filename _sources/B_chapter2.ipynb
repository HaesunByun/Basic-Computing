동일한 소스 코드를 항상 같은 방식으로 읽기 때문에 매 번 반복하기 번거로운 작업을 프로그램으로 만들어 두면 그 이후로는 계속 편하게 실행 시켜서 결과를 얻을 수 있다.

## print()
print()는 문자열, 변수, 연산의 결과를 표준출력장치인 화면에 출력하는 함수이다. 


- 출력하기 위해서는 출력하고자 하는 내용을 print() 함수의 괄호안에 넣어주면 된다.

````python
print(123)
print('hello')
print(5+8)
print('hello! ' + 'sunny')
print('hello! ''sunny')
name = 'sunny'
print(name)
print('hello! ' + name)
````

- 그런데, 아래의 코드는 print() 함수를 쓰지 않아도 결과가 나왔을 것이다. 이는 단지 코랩에서 제공해주는 기능으로 실행시키는 셀의 가장 마지막 라인을 결과로 되돌려주기 때문이다. 

````python
'Hello world'
````

- 만약 아래의 코드처럼 여러줄을 print()없이 썼다면 전체 문장이 출력되는 것이 아니라 가장 마지막 줄만 출력한다.

````python
'Hello world'
'How are you?'
'Nice to meet you'
````

- 따라서 세 줄 모두 출력하려면 아래와 같이 각각 print() 괄호안에 넣어주거나 또는 하나의 print() 괄호안에 멀티라인 스트링으로 넣어준다.

````python
print('Hello world')
print('How are you?')
print('Nice to meet you')

print('''Hello world
How are you?
Nice to meet you''')
````

- print() 괄호안에 아무런 내용없다면 한 줄 공백을 출력한다. 


````python
print('Hello')
print( )
print('Nice to meet you')
````

- 여러개의 결과를 나열하여 출력하려면 콤마(,)로 구분한다. 콤마는 문자열과 숫자를 동시에 출력할 때 매우 유용하다. 콤마를 사용하면 기본으로 한칸 띄어쓰기(공백)가 된다.

````python
print('내 나이는', 20, '살이야')
````

````python
age = 20
print('10년 후의 나이는', age+10, '살이야')
````

- 한칸 띄어쓰기를 없애고 싶거나 다른 구분자를 넣고 싶다면, 옵션 sep을 변경하면 된다. 

````python
print('내 나이는', 20, '살이야', sep='')
print(2323,18549, sep='-')
````

- f-string과 함께 사용하면 띄어쓰기, 데이터타입 등 고민 할 필요없이 수월하게 출력할 수 있다.

````python
name = "민지"
age = 20
print(f'{name}는 {age}살이야')
````

- print() 함수는 줄바꿈(\n) 기능이 포함되어 있다. 줄바꿈 기능을 바꾸고 싶다면 옵션 end를 변경하면 된다. \t는 일정한 간격으로 분리해주는 tab 기능이다.


````python
print('나','너','우리', end='^^')
print('나','너','우리', end='\t')
````

- print() 함수의 다양한 활용방법을 알고 싶다면 help()를 외쳐라.

````python
help(print)
````

## input()

input() 함수는 프로그램 실행 중에 표준입력장치인 키보드로부터 데이터를 입력받는 함수이다. 프로그램 실행중에 input() 코드를 만나면 그 아래 입력박스가 생기면서 커서가 깜박거리면서 입력을 기다린다.
그 박스안에 데이터를 입력하고 엔터를 치면된다. 

입력 받은 데이터는 일반적으로 변수에 저장한다. 
만약, 변수에 저장없이 input()만 있다면 데이터는 입력 받지만 저장하지는 않는다.

아래의 코드를 실행시켜보자.
- 'Input your name.'을 출력한다.
- 입력으로 'sunny'를 입력하면 변수 myName에 'sunny'가 저장된다.
- Welcome, sunny.가 출력된다.

```python
print ('Input your name.')
myName = input()
print (f'Welcome, {myName}.')
🚀
Input your name.
sunny
Welcome, sunny.
```

## Data type 변환
* 형 변환이 필요한 이유: Data type에 따라 지원하는 연산(기능)이 다름. 
* 예) input() 함수로 입력 받은 데이터는 문자열 처리됨. 예를 들어, 숫자 17를 입력하면 문자열 '17'이 됨. 



````python
age = input('How old are you?')
age = age + 10  
print (f'Ten years later, you will be {age} years old.') 
````

````python
age = input('Input ')
print(type(age))
age = int(age)
print(type(age))
````

#### int()
* 정수로 변환해주는 함수

````python
print(int(42))
print(int(3.5))
print(int('42'))
print(int('    42   '))
````

````python
#invalid case
print(int('3.5'))
print(int('forty-two'))
````

````python
#invalid example
num=int(input('Enter body temperature: '))
print(num)
````

````python
currentAge = int(input('몇 살이세요?'))
print (f'당신의 10년 후에 나이는 {currentAge+10} 살 입니다.')
````

#### str()
* 문자열로 변환해주는 함수

````python
print(str(42))
print(str(3.5))
print(str('42'))
print(str('3.5'))
````

#### float()
* 실수형으로 변환해주는 함수

````python
print(float(42))
print(float(3.5))
print(float('3'))
print(float('3.6'))
````

## 마무리
- ‘#’은 주석처리 기호이며, 주석은 프로그램과 소스코드에 대한 부연 설명으로 프로그램의 실행에 영향을 미치지 않는다. 
- print ( ) 함수는 표준출력장치인 스크린에 값을 출력하는 함수이며, 출력하고자 하는 내용을 괄호 안에 넣어주면 된다.
- input( ) 함수는 프로그램 실행 중에 값을 입력 받는 함수다. 입력 받은 데이터는 문자열이다.
- Data type에 따라 지원하는 연산(기능)이 다르다. 필요한 연산을 위해 Data type 변환이 필요할 수가 있다.
- int ( )함수는 문자열을 정수로 바꿀 때 주로 사용한다.
- str ( ) 함수는 정수를 문자열로 바꿀 때 주로 사용한다.
- float ( ) 함수는 문자열이나 정수를 실수로 바꿀 때 주로 사용한다.
